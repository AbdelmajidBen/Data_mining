```python
# Import des bibliothèques nécessaires
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Chargement des données depuis le fichier CSV
data = pd.read_csv('DataMeteo12 Complet.csv')
```

```python
# Extraction des caractéristiques nécessaires pour l'analyse (selon les colonnes de ton fichier)
X = data[['RAIN', 'month']].values  # Assure-toi de remplacer 'feature1' et 'feature2' par les noms des colonnes appropriées
```

```python
# Visualisation des données
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.xlabel('RAIN')
plt.ylabel('month')
plt.title('Données de Métoo')
plt.show()
```

```python
# Création de l'objet KMeans avec le nombre de clusters désiré
kmeans = KMeans(n_clusters=4)  # Remplace 4 par le nombre de clusters que tu souhaites
```

```python
# Entraînement du modèle sur les données
kmeans.fit(X)
```

```python
# Obtention des centres des clusters
centers = kmeans.cluster_centers_
```

```python
# Obtention des labels des clusters pour chaque point
labels = kmeans.labels_
```

```python
# Affichage des clusters et de leurs centres
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Clusters identifiés')
plt.show()
```
